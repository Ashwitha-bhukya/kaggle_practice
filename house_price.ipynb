```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data.drop('Unnamed: 0',inplace=True, axis = 1)
```

```python
data.head()
```

```python
ax = data.property_type.value_counts().plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i)
```

```python
ax = data.city.value_counts().plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i)
```

```python
data.city.value_counts().plot(kind='pie',autopct='%.2f')
```

```python
import plotly.express as px
```

```python
fig = px.bar(data.groupby('city',as_index=False)['price'].mean().sort_values(by='price',ascending=False).head(10), x = 'city', y = 'price', color = 'city', labels = {'city':'City Name', 'price':'Price'}, template = 'ggplot2',text = 'price',title = '<b> Average Price by City')
fig.show()
```

```python
plt.figure(figsize=(12,6))
ax = data.price.groupby(data.city).mean().sort_values(ascending=False).plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = data.price.groupby(data.property_type).mean().sort_values(ascending=False).plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

```python
px.funnel(data.groupby('city',as_index=False)['price'].count().sort_values(by='price',
                    ascending=False).head(15),y='city',x='price',
          color_discrete_sequence=['purple'],labels={'price':'count'},
          template='seaborn',title='<b> Which city has more ad posting')
```
